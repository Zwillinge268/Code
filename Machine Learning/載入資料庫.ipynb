```python
#載入資料庫
from sklearn.datasets import load_iris

iris = load_iris()
#print(iris.target_names) #分類名稱
#print(iris.target) #分類代碼
#print(iris.data) #資料
```

```python
#分割資料
from sklearn.model_selection import train_test_split

X = iris.data
Y = iris.target
train_data, test_data, train_label, test_label = train_test_split(X, Y, test_size = 0.3, random_state = 42, shuffle = True)
#shuffle 為打亂資料 ， random_state 為分割參數

print(train_data.shape)
print(train_label.shape)
print(test_data.shape)
print(test_label.shape)
#shape 陣列大小
```

```python
#建立模型(使用 SVM 演算法)
from sklearn.svm import SVC

#設定參數
clf = SVC(kernel = 'rbf', C = 1, gamma = 'auto')

#填入資料
clf.fit(train_data, train_label)

#模型正確率測試
print("訓練資料正確率", clf.score(train_data, train_label))
print("測試資料正確率", clf.score(test_data, test_label))
```



## 建立訓練模型（鳶尾花決策樹）

```python
from sklearn import tree
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split

# 讀入鳶尾花資料
iris = load_iris()
iris_data = iris.data
iris_label = iris.target

# 切分訓練與測試資料
train_data, test_data, train_label, test_label = train_test_split(iris_data, iris_label, test_size = 0.3)

# 建立分類器
clf = tree.DecisionTreeClassifier()
iris_clf = clf.fit(train_data, train_label)

print("訓練資料正確率:",clf.score(train_data,train_label)) # 訓練資料正確率
print("測試資料正確率:",clf.score(test_data, test_label)) # 測試資料正確率

predict=clf.predict(test_data) # 預測
print("標準答案:",test_label) # 標準答案
print("預測答案:",predict) # 預測答案
```

```python
#圖形化結果
import pydotplus  
from IPython.display import Image  
dot_data = tree.export_graphviz(clf, out_file=None, 
                feature_names=iris.feature_names,
                class_names=iris.target_names,  
                filled=True, rounded=True,  
                special_characters=True)  
graph = pydotplus.graph_from_dot_data(dot_data)  
Image(graph.create_png())  #將結果用圖形化呈現
```

## 建立訓練模型（鳶尾花隨機森林）

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split

# 讀入鳶尾花資料
iris = load_iris()
iris_data = iris.data
iris_label = iris.target

# 切分訓練與測試資料
train_data, test_data, train_label, test_label = train_test_split(iris_data, iris_label, test_size = 0.3)

# 建立分類器
clf = RandomForestClassifier(n_estimators=80, max_depth=4)
iris_clf = clf.fit(train_data, train_label)
print("訓練資料正確率:",clf.score(train_data,train_label)) # 訓練資料正確率
print("測試資料正確率:",clf.score(test_data, test_label)) # 測試資料正確率

predict=clf.predict(test_data) # 預測
print("標準答案:",test_label) # 標準答案
print("預測答案:",predict) # 預測答案
```

## 建立訓練模型（鳶尾花貝氏分類器）

```python
from sklearn import datasets
from sklearn.naive_bayes import GaussianNB
from sklearn.model_selection import train_test_split

# 讀入鳶尾花資料
iris=datasets.load_iris()
x=iris.data
y=iris.target

# 切分訓練與測試資料
train_data, test_data, train_label, test_label = train_test_split(x, y,test_size=0.3,random_state=0)

# 建立分類器
clf=GaussianNB()
clf.fit(train_data,train_label)
print("訓練資料正確率:",clf.score(train_data,train_label)) # 訓練資料正確率
print("測試資料正確率:",clf.score(test_data, test_label)) # 測試資料正確率

predict=clf.predict(test_data) # 預測
print("標準答案:",test_label) # 標準答案
print("預測答案:",predict) # 預測答案
```
